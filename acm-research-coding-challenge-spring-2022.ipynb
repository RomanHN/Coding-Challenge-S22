## preparing dataset

```python
# importing data

mushrooms = pd.read_csv('../input/mushroom-csv-file/mushrooms.csv')
mushrooms.head()
```

```python
# converting data to numerical format

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for column in mushrooms.columns:
    mushrooms[column] = encoder.fit_transform(mushrooms[column])

mushrooms.head()
```

```python
# splitting data into dependent and independent variables

X = mushrooms.drop('class',axis=1)
y = mushrooms['class']
X.shape
y.shape
```

```python
# splitting data into training and testing sets
## The first 70% of the mushrooms are the training set, and the last 30% will be used to test.

X_train, X_test, y_train, y_test = X[:5686], X[5686:], y[:5686], y[5686:]
```

```python
# I will use a basic Stochastic Gradient Descent model to classify the mushrooms.

from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
classifier = make_pipeline(StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3))
classifier.fit(X_train, y_train)
```

```python
# evaluating model (spoiler: it's terrible)
num_correct = 0
for x in range(2438):
    if classifier.predict([X_test.iloc[x]]) == y_test.iloc[x]:
        num_correct += 1
        
print(num_correct)
print(num_correct / 2438)
```
